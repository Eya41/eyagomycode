```python
List = list()
vals = input("Enter your values: ") 
val = vals.split(" ")
print(val)
for i in val:
    n=int(i)
    x=range (1,(n+1))
    if n<0 :
        print("Calcul de factoriel impossible!")
    elif n==0:
        List.append(1)
    else:
        fact=1
        for j in x:
            fact = fact*j
        List.append(fact)
print(List)  
```

```python
import math
i = 0
while(i != 1):
    try :
        val = input("Enter a comma-separated sequence: ") 
        D= val.split(",")
        res=""
        C=50
        H=30

        for i in D :
            Q= round(math.sqrt((2*C*int(i)/H)))
            res = res + str(Q) +","
        i = 1
    except:
        pass
print(res)
```
